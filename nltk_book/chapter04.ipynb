```python
# this function updates the content of frequency distribution that is passed in as parameter and it also prints a list of n most frequent words
from urllib import request
from bs4 import BeautifulSoup
from nltk import word_tokenize
import re
import nltk
url = 'https://www.archives.gov/founding-docs/declaration-transcript'
text = request.urlopen(url).read().decode('utf8')
soup = BeautifulSoup(text)
content = soup.get_text()
fdist = nltk.FreqDist()
for word in word_tokenize(content):
    fdist[word.lower()] += 1
result = []
for word,count in fdist.most_common():
    result.append(word)
```

```python
#url fdist n
def freq_words(url,fdist,n):
    text = request.urlopen(url).read().decode('utf8')
    soup = BeautifulSoup(text)
    content = soup.get_text()
    fdist = nltk.FreqDist() 
    for words in word_tokenize(content):
        fdist[words.lower()] += 1
    results = []
    for word,value in fdist.most_common(n):
        results.append(word)
    return results
```

```python
houses = ["Eric's house", "Kenny's house", "Kyle's house", "Stan's house"]
def deliver(houses):
    if len(houses) == 1:
        house = houses[0]
        print('Delivering Present to',house)
    else:
        mid = len(houses) // 2
        first_half = houses[:mid]
        second_half = houses[mid:]
        deliver(first_half)
        deliver(second_half)
        

    
```

```python
# doing the same computation as mentioned above but in this case we use global parameters
# Global mutable state
current_number = 1
accumulated_number = 0
def sum_recursive():
    global current_number
    global accumulated_number
    
    if current_number == 11:
        return accumulated_number
    else:
        current_number += 1
        accumulated_number = current_number + accumulated_number
        return sum_recursive()
```

```python
# calculating sum of all elements of the list recursively
def list_sum_recursive(input_list):
    # Base Case
    if input_list == []:
        return 0
    # Recursive case
    # Decompose the original problem into simple instances of same problem
    # by making use of the fact that the input is recursive data structure
    # and can be defined in terms of a smaller version of itself
    else:
        head = input_list[0]
        smaller_list = input_list[1:]
        return head + list_sum_recursive(smaller_list)
```

```python
def ls_recursive(input_list):
    if input_list == []:
        return 0
    else:
        head = input_list[0]
        small_list = input_list[1:]
        return head + ls_recursive(small_list)
```

```python
def fib_rec(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib_rec(n-1) + fib_rec(n-2)
```

```python
def insert(trie, key, value):
    if key:
        first, rest = key[0], key[1:]
        if first not in trie:
            trie[first] = {}
        insert(trie[first], rest, value)
    else:
        trie['value'] = value
```
