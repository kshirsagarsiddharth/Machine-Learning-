```python
from sklearn.model_selection import train_test_split
X_train,X_validation_test,y_train,y_validation_test = train_test_split(X,y,test_size = 0.02,random_state = 45)
X_validation,X_test,y_validation,y_test = train_test_split(X_validation_test,y_validation_test,test_size = 0.5,random_state = 45)
```

# unigrams countVectorizer

```python
if len(X_test[y_test == 0])/len(X_test) > 0.5:
    null_accuracy = len(X_test[y_test == 0])/len(X_test)
else:
    null_accuracy = 1 - len(X_test[y_test == 0])/len(X_test)
```

```python
clf = LogisticRegression(verbose = 35,n_jobs = -1)
unigram_count_vec = CountVectorizer(max_features = 50000,stop_words ='english')
unigram_pipeline = Pipeline([('vectorizer',unigram_count_vec),('classifier',clf)],verbose = True)
unigram_pipeline.fit(X_train,y_train)
```

```python
pred = unigram_pipeline.predict(X_test)
print(classification_report(pred,y_test))
print('The accuracy score is',accuracy_score(pred,y_test))
plot_confusion_matrix(pred, y_test,figsize=(7,7))
```

```python
pred = unigram_pipeline.predict(X_test)
print(classification_report(pred,y_test))
print('The accuracy score is',accuracy_score(pred,y_test))
plot_confusion_matrix(pred, y_test,figsize=(7,7))
```
