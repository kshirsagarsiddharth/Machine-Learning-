```python
df = pd.read_excel('/kaggle/input/tfidf-cvec-logistic-score-time/tfidf_cvec_logistic_score_time.xlsx')
```

```python
df.columns
```

```python
plt.figure(figsize = (10,7))
sns.set_style('dark')
sns.lineplot(data = df,x = 'features' ,
             y = 'cv_uni_score_with',
             color = 'green',
             marker = 'o',
             linewidth = 4 ,
             label = 'stopwords',
             alpha = 0.5)

sns.lineplot(data = df,x = 'features' ,
             y = 'cv_uni_score_without',
             color = 'red',
             marker = 'o',
             linewidth = 4 ,
             label = 'without stopwords',
             alpha = 0.5)
plt.xlabel('features')
plt.ylabel('accuracy score')
plt.title('Unigram with count vec')
plt.legend()
plt.show()
```
