```python
from scipy.stats import skew
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x1 = np.linspace( -5, 5, 1000 ) 
y1 = 1./(np.sqrt(2.*np.pi)) * np.exp( -.5*(x1)**2  ) 
plt.plot(x1,y1,'*')
print('Data skewness :',skew(y1))
#here the skewness of data is greater than zero this implies that this data has more weight on thhe left tail
```

```python
x1 = np.linspace( -5, 12, 1000 ) 
y1 = 1./(np.sqrt(2.*np.pi)) * np.exp( -.5*(x1)**2  ) 
plt.plot(x1,y1)
skew(y1)
#more weight on the left tail
```

```python
from scipy.stats import skew
x = np.random.normal(0,2,10000)
sns.distplot(x)
skew(x)
#here the value of skew is less than zero this impies that the data is more skewed towards the right or 
#the right tail has more weight
```

# np.log1p

```python
in_array = [1,3,5]
print(in_array)
print(np.log1p(in_array))
```

```python
in_array = [1,1.2,1.4,1.6,1.8,2]
out_array = np.log1p(in_array)
y = [1,1.2,1.4,1.6,1.8,2]
plt.plot(in_array,y,marker = '^',label = 'normal')
plt.plot(out_array,y,marker = '^',label = 'log1p')
plt.legend()
```
